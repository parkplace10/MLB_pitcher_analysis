```python
import pandas as pd
```

```python
pitch_file = "pitches.csv"
pitch_df = pd.read_csv(pitch_file)
pitch_df.head()
```

```python
atbat_file = "atbats.csv"
atbat_df = pd.read_csv(atbat_file)
atbat_df.head()
```

```python
game_file = "games.csv"
game_df = pd.read_csv(game_file)
game_df.head()
```

```python
player_file = "player_names.csv"
player_df = pd.read_csv(player_file)
player_df.head()
```

```python
pitch1_df = pitch_df.merge(atbat_df, on='ab_id', how='outer')
pitch1_df.head()
```

```python
pitch2_df = pitch1_df.merge(game_df, on='g_id', how='outer')
pitch2_df.head()
```

```python
pitch3_df = pd.merge(pitch2_df, player_df, left_on='batter_id', right_on='id')
pitch3_df.head()
```

```python
merged_df = pd.merge(pitch3_df, player_df, left_on='pitcher_id', right_on='id')
merged_df.head()
```

```python
merged_df = merged_df[['ab_id', 'b_count', 'b_score', 'break_angle','break_length', 'break_y', \
'code', 'end_speed', 'nasty', 'on_1b', 'on_2b', 'on_3b', 'outs', 'pfx_x', 'pfx_z', 'pitch_num', \
'pitch_type', 'px', 'pz', 's_count', 'spin_rate', 'start_speed', 'type', 'batter_id', 'event', \
'g_id', 'inning', 'p_score', 'p_throws', 'pitcher_id', 'stand', 'top', 'away_team', 'home_team', \
'venue_name', 'id_x', 'first_name_x', 'last_name_x', 'id_y', 'first_name_y', 'last_name_y']].copy()
merged_df.head()
```

```python
merged_df['batter_name'] = merged_df['first_name_x'].str.cat(merged_df['last_name_x'],sep=" ")
merged_df['pitcher_name'] = merged_df['first_name_y'].str.cat(merged_df['last_name_y'],sep=" ")
merged_df.head()
```

```python
cleanish_df = merged_df[['ab_id', 'b_count', 'b_score', 'break_angle','break_length', 'break_y', \
'code', 'end_speed', 'nasty', 'on_1b', 'on_2b', 'on_3b', 'outs', 'pfx_x', 'pfx_z', 'pitch_num', \
'pitch_type', 'px', 'pz', 's_count', 'spin_rate', 'start_speed', 'type', 'batter_id', 'event', \
'g_id', 'inning', 'p_score', 'p_throws', 'pitcher_id', 'stand', 'top', 'away_team', 'home_team', \
'venue_name', 'batter_name', 'pitcher_name']].copy()
cleanish_df.head()
```

```python
cleanish_df.columns = ['atbat_id', 'count_balls', 'score_batter', 'break_sharpness', 'break_amount', \
'break_location', 'pitch_result_subcat', 'speed_end', 'nasty', 'on_1b', 'on_2b', 'on_3b', 'outs', \
'movement_horizontal', 'movement_vertical', 'pitch_number', 'pitch_type', 'pitch_location_x', \
'pitch_location_y', 'count_strikes', 'pitch_spin_rate', 'speed_start', 'pitch_result', 'batter_id', \
'atbat_result', 'game_id', 'inning', 'score_pitcher', 'pitcher_hand', 'pitcher_id', 'batter_stance', \
'inning_top', 'away_team', 'home_team', 'stadium', 'batter_name', 'pitcher_name']
cleanish_df.head()
```

```python
clean_df = cleanish_df[['pitcher_id', 'pitcher_name', 'pitcher_hand', 'count_balls', 'count_strikes', \
'outs', 'pitch_number', 'pitch_type', 'pitch_result', 'pitch_result_subcat', 'atbat_result', \
'speed_start', 'speed_end', 'movement_horizontal', 'movement_vertical', 'pitch_spin_rate', \
'pitch_location_x', 'pitch_location_y', 'break_sharpness', 'break_amount', 'break_location', 'nasty', \
'batter_id', 'batter_name', 'batter_stance', 'score_pitcher', 'score_batter', 'atbat_id', \
'inning', 'inning_top', 'on_1b', 'on_2b', 'on_3b', 'game_id', 'home_team', 'away_team', 'stadium']]
clean_df.head()
```

```python
pd.options.display.max_columns=None
display(clean_df)
```

```python
clean_df['pitch_type'].replace('FF', 'Four-Seamer', inplace=True)
clean_df['pitch_type'].replace('CU', 'Curveball', inplace=True)
clean_df['pitch_type'].replace('CH', 'Changeup', inplace=True)
clean_df['pitch_type'].replace('EP', 'Eephus', inplace=True)
clean_df['pitch_type'].replace('FC', 'Cutter', inplace=True)
clean_df['pitch_type'].replace('FO', 'Pitchout', inplace=True)
clean_df['pitch_type'].replace('FS', 'Splitter', inplace=True)
clean_df['pitch_type'].replace('FT', 'Two-Seamer', inplace=True)
clean_df['pitch_type'].replace('IN', 'Intentional Ball', inplace=True)
clean_df['pitch_type'].replace('KC', 'Knuckle Curve', inplace=True)
clean_df['pitch_type'].replace('KN', 'Knuckleball', inplace=True)
clean_df['pitch_type'].replace('PO', 'Pitchout', inplace=True)
clean_df['pitch_type'].replace('SC', 'Curveball', inplace=True)
clean_df['pitch_type'].replace('SI', 'Sinker', inplace=True)
clean_df['pitch_type'].replace('SL', 'Slider', inplace=True)
clean_df['pitch_type'].replace('UN', 'Unknown', inplace=True)
clean_df['pitch_type'].replace('FA', 'X', inplace=True)
clean_df['pitch_type'].replace('AB', 'Y', inplace=True)
clean_df.head()
```

```python
clean_df['pitch_type'].value_counts()
```

```python
clean_df['pitch_result'].replace('B', 'Ball', inplace=True)
clean_df['pitch_result'].replace('S', 'Swinging Strike', inplace=True)
clean_df['pitch_result'].replace('X', 'In Play, Out', inplace=True)
clean_df.head()
```

```python
clean_df['pitch_result_subcat'].replace('B', 'Ball', inplace=True)
clean_df['pitch_result_subcat'].replace('*B', 'Ball', inplace=True)
clean_df['pitch_result_subcat'].replace('C', 'Called Strike', inplace=True)
clean_df['pitch_result_subcat'].replace('D', 'In Play, No Out', inplace=True)
clean_df['pitch_result_subcat'].replace('E', 'In Play, Runs', inplace=True)
clean_df['pitch_result_subcat'].replace('F', 'Foul Ball', inplace=True)
clean_df['pitch_result_subcat'].replace('H', 'Hit By Pitch', inplace=True)
clean_df['pitch_result_subcat'].replace('I', 'Intentional Ball', inplace=True)
clean_df['pitch_result_subcat'].replace('L', 'Foul Bunt', inplace=True)
clean_df['pitch_result_subcat'].replace('M', 'Missed Bunt', inplace=True)
clean_df['pitch_result_subcat'].replace('P', 'Pitchout', inplace=True)
clean_df['pitch_result_subcat'].replace('Q', 'Swinging Pitchout', inplace=True)
clean_df['pitch_result_subcat'].replace('R', 'Foul Pitchout', inplace=True)
clean_df['pitch_result_subcat'].replace('S', 'Swinging Strike', inplace=True)
clean_df['pitch_result_subcat'].replace('T', 'Foul Tip', inplace=True)
clean_df['pitch_result_subcat'].replace('W', 'Swinging Strike', inplace=True)
clean_df['pitch_result_subcat'].replace('X', 'In Play, Out', inplace=True)
clean_df['pitch_result_subcat'].replace('V', 'V', inplace=True)
clean_df['pitch_result_subcat'].replace('Z', 'Z', inplace=True)
clean_df.head()
```

```python
clean_df = clean_df[clean_df.pitch_type != 'X']
clean_df = clean_df[clean_df.pitch_type != 'Y']
clean_df = clean_df[clean_df.pitch_type != 'Unknown']
clean_df = clean_df[clean_df.pitch_result_subcat != 'V']
clean_df = clean_df[clean_df.pitch_result_subcat != 'Z']
```
