```python
def get_pitch_location_category(row):
    #check high left
    if row['pitch_location_y'] > 3.6 and row['pitch_location_x']<-1:
        return "high and left"
    #check high right
    elif row['pitch_location_y'] > 3.6 and row['pitch_location_x']>1:
        return "high and right"
    #check low left
    elif row['pitch_location_y'] < 1.6 and row['pitch_location_x']<-1:
        return "low and left"
    #check low right
    elif row['pitch_location_y'] < 1.6 and row['pitch_location_x']>1:
        return "low and right"
    #check high
    elif row['pitch_location_y'] > 3.6:
        return "high"
    #check low
    elif row['pitch_location_y'] < 1.6:
        return "low"
    #check left
    elif row['pitch_location_x'] <-1:
        return "left"
    #check right
    elif row['pitch_location_x'] >1:
        return "right"
    #check top zone
    elif row['pitch_location_y'] >= 2.93337:
        if row['pitch_location_x'] >=-1 and row['pitch_location_x'] <-.3333:
            return "top left"
        elif row['pitch_location_x'] >=-.3333 and row['pitch_location_x'] <.3333:
            return "top center"
        elif row['pitch_location_x'] >=.3333:
            return "top right"
    #check middle zone
    elif row['pitch_location_y'] >= 2.26667:
        if row['pitch_location_x'] >=-1 and row['pitch_location_x'] <-.3333:
            return "middle left"
        elif row['pitch_location_x'] >=-.3333 and row['pitch_location_x'] <.3333:
            return "middle center"
        elif row['pitch_location_x'] >=.3333:
            return "middle right"
    #check bottom zone
    elif row['pitch_location_y'] >= 1.6:
        if row['pitch_location_x'] >=-1 and row['pitch_location_x'] <-.3333:
            return "bottom left"
        elif row['pitch_location_x'] >=-.3333 and row['pitch_location_x'] <.3333:
            return "bottom center"
        elif row['pitch_location_x'] >=.3333:
            return "bottom right"
    
```

```python
clean_df['pitch_location_category'] = clean_df.apply(get_pitch_location_category, axis=1)
```
